```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
p1 = '/home/mehul/Documents/lectures/python/data/movielens/movies.dat'
c1 = ['movie_id', 'title', 'genre']
movies = pd.read_table(p1, delimiter='::', names=c1, header=None)
movies[:5]
```

```python
p2 = '/home/mehul/Documents/lectures/python/data/movielens/users.dat'
c2 = ['user_id', 'gender', 'age_group', 'occupation', 'zip']
users = pd.read_table(p2, delimiter='::', names=c2, header=None)
users[:10]
```

```python
p3 = '/home/mehul/Documents/lectures/python/data/movielens/ratings.dat'
c3 = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table(p3, delimiter='::', names=c3, header=None)
ratings
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data
```

```python
genderized_mean = data.pivot_table('rating', aggfunc=np.mean, index='title', columns='gender')
c_d = genderized_mean.fillna(0)
c_d
```

```python
female_domination = c_d.sort_values(by='F', ascending=False)
female_domination
```

```python
data[:10]
```

```python
title_grouped = data.groupby('title')
ratings_by_title = title_grouped['rating'].count()
popular_films = ratings_by_title[ratings_by_title >= 250]
popular_film_names = popular_films.index
popular_film_names
```

```python
popular_means = c_d.loc[popular_film_names]
popular_means.sort_values(by='F', ascending=False)[:10].plot(kind='barh')
plt.show()
```

```python
popular_means['Diff'] = popular_means['F'] - popular_means['M']
popular_means.sort_values(by='Diff', ascending=False)[:10]
```
