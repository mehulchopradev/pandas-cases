```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
import json
```

```python
path = '/home/mehul/Documents/lectures/python/data/url-shortner/usagov_bitly_data2012-03-16-1331923249.txt'
url_file = open(path)
data = [json.loads(line) for line in url_file]
url_file.close()
data
```

```python
df = DataFrame(data)
df
```

```python
c1_d = df['tz'].fillna('Missing')
c1_d[c1_d == ''] = 'Unknown'
c1_d
```

```python
c1_d.value_counts()[:10].plot(kind='barh')
plt.show()
```

```python
os_types = np.where(df['a'].str.contains('Windows'), 'Windows', 'Non-Windows')
# print os_types
# print type(os_types)
os_series = Series(os_types)
os_series
```

```python
df['os'] = os_series
p_df = df[['a', 'tz', 'os']]
c_p_df = p_df.fillna('Missing')
c_p_df[c_p_df == ''] = 'Unknown'
c_p_df
```

```python
g_data = c_p_df.groupby(['tz', 'os'])
win_non_win = g_data.count().unstack().fillna(0)
top_ten = win_non_win.sum(axis=1).sort_values(ascending=False)[:10].index
win_non_win.loc[top_ten].plot(kind='barh', stacked=True)
plt.show()
```
